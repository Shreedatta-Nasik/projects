```python
import rotate_matrix
import numpy
```

```python
top_face=[]
front_face=[]
back_face=[]
left_face=[]
bottom_face=[]
right_face=[]
```

```python
def input_cube():
    global front_face,right_face,back_face,left_face,top_face,bottom_face
    cube_code={"R":1,"B":2,"G":3,"O":4,"Y":5,"W":6}
    def input_face(face):
        print("Enter the matrix of ")
        m=[]
        for i in range(3):
            a=[]
            for j in range(3):
                t=input()
                a.append(cube_code.get(t.upper))
            m.append(a)
        return m

    front_face=input_face(front_face)
    right_face=input_face(right_face)
    back_face=input_face(back_face)
    left_face=input_face(left_face)
    top_face=input_face(top_face)
    bottom_face=input_face(bottom_face)
    return 0
```

```python
def up_right():
    global front_face,right_face,left_face,back_face,bottom_face,top_face
    temp=numpy.array(front_face)[:,2]
    front_face[0][2]=bottom_face[0][2]
    front_face[1][2]=bottom_face[1][2]
    front_face[2][2]=bottom_face[2][2]
    bottom_face[0][2]=back_face[2][0]
    bottom_face[1][2]=back_face[1][0]
    bottom_face[2][2]=back_face[0][0]
    back_face[0][0]=top_face[2][2]
    back_face[1][0]=top_face[1][2]
    back_face[2][0]=top_face[0][2]
    top_face[0][2]=temp[0]
    top_face[1][2]=temp[1]
    top_face[2][2]=temp[2] 
    m=[]
    m=rotate_matrix.anti_clockwise(right_face)
    right_face[0]=list(m[0])
    right_face[1]=list(m[1])
    right_face[2]=list(m[2])
    return 0
```

```python
def down_right():
    global front_face,right_face,left_face,back_face,bottom_face,top_face
    up_right()
    up_right()
    up_right()
    return 0
```

```python
def up_left():
    global front_face,right_face,left_face,back_face,bottom_face,top_face
    temp=numpy.array(front_face)[:,0]
    front_face[0][0]=bottom_face[0][0]
    front_face[1][0]=bottom_face[1][0]
    front_face[2][0]=bottom_face[2][0]
    bottom_face[0][0]=back_face[2][2]
    bottom_face[1][0]=back_face[1][2]
    bottom_face[2][0]=back_face[0][2]
    back_face[0][2]=top_face[2][0]
    back_face[1][2]=top_face[1][0]
    back_face[2][2]=top_face[0][0]
    top_face[0][0]=temp[0]
    top_face[1][0]=temp[1]
    top_face[2][0]=temp[2] 
    m=[]
    m=rotate_matrix.clockwise(left_face)
    left_face[0]=list(m[0])
    left_face[1]=list(m[1])
    left_face[2]=list(m[2])
    return 0
```

```python
def down_left():
    global front_face,right_face,left_face,back_face,bottom_face,top_face
    up_left()
    up_left()
    up_left()
    return 0
```

```python
def left():
    global front_face,right_face,left_face,back_face,bottom_face,top_face
    temp=front_face[0]
    front_face[0]=left_face[0]
    left_face[0]=back_face[0]
    back_face[0]=right_face[0]
    right_face[0]=temp
    m=[]
    m=rotate_matrix.clockwise(top_face)
    top_face[0]=list(m[0])
    top_face[1]=list(m[1])
    top_face[2]=list(m[2])
    return 0  
```

```python
def right():
    global front_face,right_face,left_face,back_face,bottom_face,top_face
    temp=front_face[0]
    front_face[0]=right_face[0]
    right_face[0]=back_face[0]
    back_face[0]=left_face[0]
    left_face[0]=temp
    m=[]
    m=rotate_matrix.anti_clockwise(top_face)
    top_face[0]=list(m[0])
    top_face[1]=list(m[1])
    top_face[2]=list(m[2])
    return 0
```

```python
def front_right():
    global front_face,right_face,left_face,back_face,bottom_face,top_face
    temp=numpy.array(left_face)[:,2]
    left_face[0][2]=bottom_face[0][0]
    left_face[1][2]=bottom_face[0][1]
    left_face[2][2]=bottom_face[0][2]
    bottom_face[0][0]=right_face[2][0]
    bottom_face[0][1]=right_face[1][0]
    bottom_face[0][2]=right_face[0][0]
    right_face[0][0]=top_face[2][0]
    right_face[1][0]=top_face[2][1]
    right_face[2][0]=top_face[2][2]
    a=temp[2]
    b=temp[1]
    c=temp[0]
    top_face[2][0]=a
    top_face[2][1]=b
    top_face[2][2]=c
    m=[]
    m=rotate_matrix.anti_clockwise(front_face)
    front_face[0]=list(m[0])
    front_face[1]=list(m[1])
    front_face[2]=list(m[2])
    return 0
```

```python
def front_left():
    global front_face,right_face,left_face,back_face,bottom_face,top_face
    temp=bottom_face[0]
    a=temp[2]
    b=temp[1]
    c=temp[0]
    bottom_face[0][0]=left_face[0][2]
    bottom_face[0][1]=left_face[1][2]
    bottom_face[0][2]=left_face[2][2]
    left_face[0][2]=top_face[2][2]
    left_face[1][2]=top_face[2][1]
    left_face[2][2]=top_face[2][0]
    top_face[2][0]=right_face[0][0]
    top_face[2][1]=right_face[1][0]
    top_face[2][2]=right_face[2][0]
    right_face[0][0]=a
    right_face[1][0]=b
    right_face[2][0]=c
    m=[]
    m=rotate_matrix.clockwise(front_face)
    front_face[0]=list(m[0])
    front_face[1]=list(m[1])
    front_face[2]=list(m[2])
    return 0
```

```python
def rotate_cube():
    global front_face,right_face,left_face,back_face,bottom_face,top_face
    temp=front_face
    front_face=right_face
    right_face=back_face
    back_face=left_face
    left_face=temp
    m=[]
    m=rotate_matrix.anti_clockwise(top_face)
    top_face[0]=list(m[0])
    top_face[1]=list(m[1])
    top_face[2]=list(m[2])
    m1=[]
    m1=rotate_matrix.clockwise(bottom_face)
    bottom_face[0]=list(m1[0])
    bottom_face[1]=list(m1[1])
    bottom_face[2]=list(m1[2])
    return 0
```
